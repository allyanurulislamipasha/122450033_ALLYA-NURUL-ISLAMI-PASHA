Data yang saya gunakan adalah dataset jumlah penduduk Kota Bandung yang mencakup beberapa kolom, yaitu :
*   Nama kecamatan dan kelurahan
*   Jenis kelamin
*   Jumlah penduduk
*   Tahun dan semester pendataan


Tahun dan semester pendataan
Saya akan menggunakan kolom:
* bps_nama_kecamatan: Nama kecamatan
* jenis_kelamin: Jenis kelamin penduduk
* jumlah_penduduk: Jumlah penduduk


**Langkah 2: Menentukan Tujuan, User, Fungsi, dan Tone**
* Tujuan: Visualisasi bertujuan untuk memberikan gambaran jumlah penduduk berdasarkan kecamatan dan jenis kelamin di Kota Bandung.
* User: Pihak-pihak terkait dalam pengambilan keputusan kebijakan pemerintah, peneliti demografi, dan warga yang ingin mengetahui distribusi populasi.
* Fungsi: Sebagai alat bantu untuk memahami perbedaan jumlah penduduk laki-laki dan perempuan di setiap kecamatan, sehingga dapat digunakan untuk perencanaan dan alokasi sumber daya.
*  Tone: Informasional dan netral, fokus pada penyajian data secara visual dengan penekanan pada keterbacaan dan kejelasan.







```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
file_path = '/content/jumlah penduduk bandung.csv'
data = pd.read_csv(file_path)

# Filter dan mempersiapkan data
data_filtered = data[['bps_nama_kecamatan', 'jenis_kelamin', 'jumlah_penduduk']].copy()

# mengelompokkan dan menjumlahkan data berdasarkan kecamatan dan jenis kelamin
data_grouped = data_filtered.groupby(['bps_nama_kecamatan', 'jenis_kelamin']).sum().reset_index()

# menentukan ukuran plot
plt.figure(figsize=(12, 8))

# membuat grafik batang menggunakan Seaborn
sns.barplot(x='bps_nama_kecamatan', y='jumlah_penduduk', hue='jenis_kelamin', data=data_grouped)

# menambahkan judul dan label
plt.title('Jumlah Penduduk Berdasarkan Jenis Kelamin di Setiap Kecamatan Kota Bandung (2017)', fontsize=16)
plt.xlabel('Nama Kecamatan', fontsize=12)
plt.ylabel('Jumlah Penduduk', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Jenis Kelamin')

# menampilkan plot
plt.tight_layout()
plt.show()
```
